```python
import re
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime

MicrosoftResearch = ""
MckinseyResearch = ""

data=[]

search_strings = ["MO", ]

links=[]

for srch in search_strings:
    searchlinks = "https://finance.yahoo.com/quote/MO/history?p=" + srch
    links.append(searchlinks)
    print(searchlinks)
print(links)


for currentpage in links:
    print("**Now Scraping:",currentpage)
    
    currentpage_resp=requests.get(currentpage)
    
    currentpagesoup=BeautifulSoup(currentpage_resp.content,"html.parser")
    
    print(currentpagesoup)
    
    articles=currentpagesoup.find_all('a')
    
    print(articles)
```
