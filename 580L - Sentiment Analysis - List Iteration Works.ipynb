```python
import numpy as np
import pandas as pd
import re
import nltk
#nltk.download('punkt')
import matplotlib.pyplot as plt

#!pip install textblob
#!pip install newspaper3k

# The punkt download enabled nlp() as a function to be done. 

articles = pd.read_csv('Articles.csv')
#print(articles.head(10))

urls = []

testurls = ['https://www.vox.com/polyarchy/2016/12/9/13898340/democracy-warning-signs', 'https://www.businessinsider.com/colts-gm-ryan-grigson-andrew-luck-contract-2016-10', 'https://www.reuters.com/article/us-davos-meeting-trump-mueller/trump-denies-report-he-ordered-mueller-fired-idUSKBN1FF12A']

for url in articles['url']:
    urls.append(url)
print(urls)

```

```python
from textblob import TextBlob
from newspaper import Article

sentimentscores = []

for urled in testurls:
    article = Article(urled)
    article.download()
    article.parse()
    article.nlp()
    text = article.summary
    blob = TextBlob(text)
    sentiment = blob.sentiment.polarity
    sentimentscores.append(sentiment)
print(sentimentscores)


#test URLs work
```
